```python
import pandas as pd
from datetime import datetime
import os

# Nombre de nuestro archivo de base de datos
ARCHIVO_DATOS = 'tracking_electoral.csv'

def cargar_datos():
    """
    Carga el DataFrame desde el archivo CSV.
    Si el archivo no existe, crea un DataFrame vacío con las columnas necesarias.
    """
    if os.path.exists(ARCHIVO_DATOS):
        # Carga los datos y se asegura de que la columna 'fecha' se lea como fecha
        try:
            df = pd.read_csv(ARCHIVO_DATOS, parse_dates=['fecha'])
        except pd.errors.EmptyDataError:
            # Si el archivo existe pero está vacío
            return crear_df_vacio()
    else:
        df = crear_df_vacio()

    return df

def crear_df_vacio():
    """Crea un DataFrame vacío con la estructura definida."""
    columnas = ['fecha', 'candidato', 'intencion_voto', 'fuente']
    return pd.DataFrame(columns=columnas)

def agregar_medicion(candidato, intencion_voto, fuente):
    """
    Agrega una nueva medición (una nueva encuesta) al tracking.

    Args:
        candidato (str): Nombre del candidato.
        intencion_voto (float): Porcentaje de intención de voto (ej: 42.5).
        fuente (str): Nombre de la encuestadora (ej: "Encuestadora X").
    """
    # Carga la base de datos actual
    df = cargar_datos()

    # Obtiene la fecha de hoy (puedes cambiar esto para ingresar fechas manualmente)
    fecha_hoy = datetime.now().date()

    # Crea la nueva fila como un diccionario
    nueva_medicion = {
        'fecha': fecha_hoy,
        'candidato': candidato,
        'intencion_voto': intencion_voto,
        'fuente': fuente
    }

    # Añade la nueva fila al DataFrame
    # pd.concat es más eficiente que .append (que está obsoleto)
    df_nueva_fila = pd.DataFrame([nueva_medicion])
    df_actualizado = pd.concat([df, df_nueva_fila], ignore_index=True)

    # Guarda el DataFrame actualizado de nuevo en el CSV
    # index=False evita que se guarde una columna extra con el índice
    df_actualizado.to_csv(ARCHIVO_DATOS, index=False)

    print(f"-> Medición agregada: {candidato} con {intencion_voto}% (Fuente: {fuente})")

def mostrar_tracking_candidato(candidato):
    """Muestra la evolución histórica de un candidato específico."""
    df = cargar_datos()

    # Filtra el DataFrame para obtener solo los datos del candidato
    df_candidato = df[df['candidato'] == candidato].copy()

    if df_candidato.empty:
        print(f"No hay datos para el candidato: {candidato}")
        return

    # Ordena los datos por fecha para ver la evolución
    df_candidato = df_candidato.sort_values(by='fecha')

    print(f"\n--- Evolución de {candidato} ---")
    print(df_candidato)
    return df_candidato

# --- Ejemplo de cómo usar el script ---
if __name__ == "__main__":

    print("Iniciando sistema de tracking electoral...")

    # Simulamos agregar algunas mediciones
    # (En un uso real, llamarías a esta función cada vez que tengas una nueva encuesta)
    agregar_medicion(candidato='Candidato A', intencion_voto=42.5, fuente='Encuestadora X')
    agregar_medicion(candidato='Candidato B', intencion_voto=38.0, fuente='Encuestadora X')

    # Simulamos otra medición unos días después
    # (Para simular, volvemos a llamar. En la vida real, solo ejecutarías esto en días diferentes)
    agregar_medicion(candidato='Candidato A', intencion_voto=43.1, fuente='Encuestadora Y')
    agregar_medicion(candidato='Candidato B', intencion_voto=37.5, fuente='Encuestadora Y')

    # Mostramos la evolución de un candidato
    mostrar_tracking_candidato('Candidato A')
    mostrar_tracking_candidato('Candidato B')


```
